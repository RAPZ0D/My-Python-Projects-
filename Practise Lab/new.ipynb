### Columns check 

Table Track 

```python
import pandas as pd
from fuzzywuzzy import fuzz

# Text names to check for similarity
text_names = [
    'track id w',
    'track name',
    'track artist',
    'track popularity',
    'track album id',
    'danceability',
    'energy',
    'key',
    'loudness',
    'mode',
    'speechiness',
    'acousticness',
    'instrumentalness',
    'liveness',
    'valence',
    'tempo',
    'duration ms'
]

# Assuming 'df' is your DataFrame containing the data

# Dictionary to store similar columns for each text name
similar_columns = {}

# Iterate through each text name
for text_name in text_names:
    max_similarity = 0
    similar_column = None

    # Iterate through columns in the DataFrame
    for column in df.columns:
        similarity = fuzz.partial_ratio(str(text_name).lower(), str(column).lower())
        if similarity > max_similarity:
            max_similarity = similarity
            similar_column = column

    if similar_column is not None:
        similar_columns[text_name] = {
            'similar_column': similar_column,
            'dtype': df[similar_column].dtype,
            'similarity_score': max_similarity
        }
    else:
        similar_columns[text_name] = {
            'similar_column': None,
            'dtype': None,
            'similarity_score': 0
        }

# Print the results
for text_name, details in similar_columns.items():
    print(f"Text: '{text_name}'")
    if details['similar_column'] is not None:
        print(f"Most similar column: '{details['similar_column']}'")
        print(f"Dtype of the similar column: {details['dtype']}")
        print(f"Similarity score: {details['similarity_score']}%")
    else:
        print("No similar column found.")
    print("--------------------")
```

# Inserting values in the tables 

Track 

```python
import mysql.connector
import pandas as pd

# Database connection details
db_config = {
    'host': 'localhost',
    'user': 'root',
    'password': 'Duke#7539',
    'database': 'makeuplab'
}

# Read CSV into DataFrame
data = pd.read_csv('new_make_up_lab_data.csv')

# Extract data relevant to the 'Track' table
track_data = data[['track_id', 'track_name', 'track_artist', 'track_popularity', 'track_album_id',
                   'danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness',
                   'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo', 'duration_ms']]

# Establish a connection to the database
connection = mysql.connector.connect(
    host=db_config['host'],
    user=db_config['user'],
    password=db_config['password'],
    database=db_config['database']
)

try:
    cursor = connection.cursor()

    # Insert data into the 'Track' table
    for _, row in track_data.iterrows():
        sql = "INSERT INTO Track (track_id, track_name, track_artist, track_popularity, track_album_id, " \
              "danceability, energy, `key`, loudness, mode, speechiness, acousticness, instrumentalness, " \
              "liveness, valence, tempo, duration_ms) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, " \
              "%s, %s, %s, %s, %s)"
        values = (
            row['track_id'], row['track_name'], row['track_artist'], int(row['track_popularity']),
            row['track_album_id'], float(row['danceability']), float(row['energy']), int(row['key']),
            float(row['loudness']), int(row['mode']), float(row['speechiness']), float(row['acousticness']),
            float(row['instrumentalness']), float(row['liveness']), float(row['valence']), float(row['tempo']),
            int(row['duration_ms'])
        )
        cursor.execute(sql, values)

    # Commit the changes to the 'Track' table
    connection.commit()

    print("Data inserted successfully into the 'Track' table!")

finally:
    # Close the connection
    connection.close()
```
